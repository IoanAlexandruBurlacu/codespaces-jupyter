Pendenza

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definizione della funzione

$f(x)=x^2-2x-3$

```python
def f(x):
    return(x**3-2*x**2-x-3)
```

Verifichiamo se funziona

```python
print(f(0.5))
```

Definiamo un vettore che contenga i valori di $x$ da -10 fino a 10 ogni 0.1, sono 201 valori

```python
x=np.linspace(-10,10,201)
print(x)
```

Calcoliamo il vettore $y$ delle immagini

```python
y=f(x)
```

Scegliamo un valore dx per $\Delta x$

```python
dx=0.00001
```

Calcoliamo la pendenza di $f(x)$ per ogni $x$

```python
m=(f(x+dx)-f(x))/dx
print(m)
```

Rappresentiamo graficamente la funzione $f(x)$ e la sua pendenza $m(x)$

```python
griglia=np.linspace(-10,10,21)

plt.figure(figsize=(8,8))  #grandezza immgini
plt.plot(x,y)
plt.plot(x,m)
plt.axis([-10,10,-10,10])
plt.gca().set_aspect("equal")     # scale x e y uguali
plt.xticks(griglia)               # griglia sulle ascisse
plt.yticks(griglia)               # griglia sulle ordinate
plt.axhline(0, color='black', linewidth=1)   # aggiunge una linea in corrispondenza di y = 0
plt.axvline(0, color='black', linewidth=1)   # aggiunge una linea in corrispondenza di x = 0
plt.grid()   # aggiunge la griglia
plt.show()   # mostra il diagramma
```

Eseguiamo una regressione polinomiale su $m(x)$

```python
np.polyfit(x,m,1)
```

quindi m(x)=2x-1.99
